#### Q1– Player Record Conversion (Dictionary Format)

```python
players = []

for row in players_data:
    parts = row.split(",")  

    player_info = {}         

    for oneplayer in parts:
        pair = oneplayer.split(":")   
        key = pair[0]            
        value = pair[1]          

        if key == "avg":
            value = float(value)
        elif key == "strike_rate":
            value = float(value)
        elif key == "fitness":
            value = int(value)

        player_info[key] = value  

    players.append(player_info)   

print(players)
```

#### Q2 – Count Total Players:

```python
print("Total Players:", len(players))
```

#### Q3 – Role-wise Fitness Summary

```python
fitness_by_role = {}   
count_by_role = {}     

for player in players:
    role = player["role"]

    if role not in fitness_by_role:
        fitness_by_role[role] = 0
    if role not in count_by_role:
        count_by_role[role] = 0

    fitness_by_role[role] = fitness_by_role[role] + player["fitness"]

    count_by_role[role] = count_by_role[role] + 1

print("Average Fitness by Role:")

for role in fitness_by_role:
    total_fitness = fitness_by_role[role]
    total_players = count_by_role[role]

    avg_fitness = total_fitness / total_players
    avg_fitness = format(avg_fitness, ".2f")

    print(role + ": " + str(avg_fitness))
```
